```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# Ambil Data BBCA 1 Tahun
# ==============================
ticker = "BBCA.JK"
data = yf.download(ticker, period="1y")

# Jika kolom berbentuk MultiIndex (ada level ticker), ambil level pertama saja
if isinstance(data.columns, pd.MultiIndex):
    data.columns = data.columns.get_level_values(0)

# Reset index supaya Date jadi kolom biasa (lebih rapi saat print)
data = data.reset_index()

# Print 5 baris pertama
print(data.head())

# Pakai harga penutupan
prices = data['Close']

# Plot harga penutupan saham BBCA
plt.figure()
plt.plot(data['Date'], prices)
plt.title("Harga Penutupan BBCA")
plt.xlabel("Tanggal")
plt.ylabel("Harga")
plt.xticks(rotation=45)
plt.show()
```

```python
# Perhitungan Log Return
log_return = np.log(prices / prices.shift(1)).dropna()
```

```python
import yfinance as yf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ==============================
# PARAMETER SIMULASI
# ==============================
T = 1          # horizon waktu (1 tahun)
L = 252        # jumlah langkah waktu (hari trading)
dt = T / L     # panjang interval waktu
S0 = prices.iloc[-1].item() # Harga awal simulasi

# ==============================
# Hitung Variansi log return
# ==============================
var_log = log_return.var()

# Estimasi sigma sesuai formula
sigma = (1/np.sqrt(dt)) * np.sqrt(var_log)

# ==============================
# Hitung mean log return
# ==============================
mean_log = log_return.mean()

# Estimasi mu sesuai formula
mu = (1/dt)*mean_log + 0.5*sigma**2

```

```python
# ==============================
# SIMULASI LINTASAN GBM
# ==============================

# Pastikan mu dan sigma bertipe float
mu = float(mu)
sigma = float(sigma)

# jumlah lintasan simulasi
M = 1

# Generate random shock Z_t ~ N(0,1)
Z = np.random.normal(0, 1, (L, M))

# Inisialisasi matriks harga
S = np.zeros((L+1, M))
S[0, :] = S0   # harga awal di waktu t=0

# Iterasi simulasi
for t in range(L):
    S[t+1, :] = S[t, :] * np.exp(
        (mu - 0.5 * sigma**2) * dt
        + sigma * np.sqrt(dt) * Z[t, :]
    )

# ==============================
# PLOT HASIL SIMULASI
# ==============================

time = np.linspace(0, T, L+1)

plt.figure()
plt.plot(time, S)
plt.title("Simulasi Lintasan Harga Saham BBCA")
plt.xlabel("Waktu (Tahun)")
plt.ylabel("Harga")
plt.grid(True)
plt.show()
```
